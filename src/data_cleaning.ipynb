```python
import pandas as pd
import matplotlib.pyplot as plt

# 加载数据
df = pd.read_csv("vcdb.csv", low_memory=False)
df.describe()
```

```python
# 查看数据
# Create a DataFrame from the dtypes
dtypes_df = df.dtypes.reset_index()
dtypes_df.columns = ['Column', 'DataType']

# Output to CSV
dtypes_df.to_csv("columns_and_dtypes.csv", index=False)
```

```python

# 受害者行业分析
attack_counts = df["victim.industry"].value_counts().head(10)
plt.figure(figsize=(10, 6))
attack_counts.plot(kind="bar", title="Top 10 victim industry Types")
plt.xlabel("Industry Type")
plt.ylabel("Count")
plt.show()
```

```python

# 按行业分析事件数量
industry_counts = df["victim.industry"].value_counts().head(5)
industry_counts.plot(kind="pie", autopct="%1.1f%%", title="Incidents by Industry")
plt.show()
```
